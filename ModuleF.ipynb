# Numerical Differentiation Problem Statement

some interval. The spacing or step size of a numerical grid is the distance between adjacent points on the grid.

# Finite Difference Approximating Derivatives

The derivative of f(x) at x=a is f'(a) = lim x -> a (f(x)-f(a))/(x-a)

The forward difference estimates the slope of the function at x_j: f'(x_j) = f(x_(j+1)) - f(x_j) / (x_(j+1) - x_j)

The backward difference estimates the slope of the function at x_j: f'(x_j) = f(x_j) - f(x_(j-1)) / (x_j - x_(j-1))

The central difference estimates the slope of the function at x_j: f'(x_j) = f(x_(j+1)) - f(x_(j-1)) / (x_(j+1) - x_(j-1))

These can be approximated with Taylor series.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)
# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = -np.sin(x_diff)
# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, label = 'Exact solution')
plt.legend()
plt.show()
# Compute max error between numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

We can see above that the error is extremely small.
