# Introduction

A vector is an n-tuple, such as [2,4,7] and [7,13,78,-25].

We can create vectors with numpy:

```python
import numpy as np

vector = np.array([[7, 13, 78, 25]])
print(vector)
```

# Transpositions

A row vector is a vector of shape (1, n). A column vector is a vector of shape (n, 1).

A vector can be transposed (so a row vector becomes a column vector and vice versa):

```python
vector1 = np.array([[1, 2, 3, 4]])
print(vector1.shape)
print(vector1.T)
print(vector1.T.shape)
```

```python
vector2 = np.array([[1], [2], [3], [4]])
print(vector2.shape)
print(vector2.T)
print(vector2.T.shape)
```

# Linear Combinations

Vectors can be multiplied by scalars and added together to create linear combinations. For example, 2*[3,7,-1] + 7*[1,-4,0] = [2\*3 + 7\*1, 2\*7 + 7\*-4, 2\*-1 + 7\*0] = [13, -14, -2]

```python
v1 = np.array([[3, 7, -1]])
v2 = np.array([[1, -4, 0]])
print(2*v1 + 7*v2)
```

# Dot Product and Cross Product

A dot product is the sum of respective elements in each vector. A cross product generates a vector which is perpendicular (orthogonal) to both input vectors. These operations can also be used to find the angle between two vectors.

```python
from numpy import arccos, dot
from numpy.linalg import norm

v = np.array([[10, 9, 3]])
w = np.array([[2, 5, 12]])
print(dot(v, w.T))

theta = arccos(dot(v, w.T)/(norm(v)*norm(w)))
print(theta)
print(np.cross(v, w))
```

# Orthogonality

Two vectors are orthogonal if their dot product is zero.

```python
v = np.array([[3, 4]])
w = np.array([[-4, 3]])
print(dot(v, w.T))
```

# Matrices

```python
P = np.array([[1, 7], [2, 3], [5, 0]])
Q = np.array([[2, 6, 3, 1], [1, 2, 3, 4]])
print(P)
print(Q)
```

# Eigenvalues and Eigenvectors

An eigenvalue of a matrix A is a value lambda such that A*x = lambda*x. The vector x is called an eigenvector.

These can be calculated with Python:

```python
from numpy.linalg import eig

a = np.array([[2, 2, 4],
              [1, 3, 5],
              [2, 3, 4]])
w,v=eig(a)
print('E-value:', w)
print('E-vector', v)
```

# Linear Regression

Given a set of independent data points, an estimation function can be generated that describes the data as well as possible. Least squares regression attempts to find the parameters of the estimation function that minimize the total squared error, defined by E = the sum from i=1 to i=m of (y_hat - y_i)^2. This can be done with Python:

```python
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 101)
y = 1 + x + x * np.random.random(len(x))
A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)

plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We can see above that the red line fits the points as well as possible.
