# Continuity and Differentiation

When a function approaches a value, it is known as a limit.

lim x->a f(x) = f(a)

```python
def func(x):
  return 2*x + 5

print('Value: ' + str(func(3)))
print('Left Limit')
print('Value at 2.9: '+ str(func(2.9)))
print('Value at 2.99: '+ str(func(2.99)))
print('Value at 2.999: '+ str(func(2.999)))
print('Right Limit')
print('Value at 3.1: '+ str(func(3.1)))
print('Value at 3.01: '+ str(func(3.01)))
print('Value at 3.001: '+ str(func(3.001)))
```

We can see above that the limit is 11 for both left and right.

The derivative of a function of a real variable measures the sensitivity to
change of the function value with respect to another variable.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def der_f(x):
    return 2*x

x = np.linspace(-7, 7, 100)
y = f(x)
y_prime = der_f(x)

x_tangent = 3
y_tangent = f(x_tangent)
slope_tangent = der_f(x_tangent)

tangent_line = slope_tangent * (x - x_tangent) + y_tangent

plt.figure(figsize=(8, 6))

plt.plot(x, y, label=r'$f(x) = x^2$', color='blue')
plt.plot(x, tangent_line, label=r'Tangent Line at $x=3$', color='red')

plt.title('Plot of $x^2$ and its Tangent Line at $x=3$')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)
plt.show()
```

We can see that x^2 and the tangent line have the same slope at x=3.

# Taylor's Theorem

Taylor’s Theorem gives an approximation of a differentiable function around
a given point by a polynomial.

In the case of the function e^x, its approximation is x^n/n! for n from 0 to the number of terms of the approximation.

```python
def exp_function(x):
    return np.exp(x)

def taylor_series(x):
    taylor_sum = np.zeros_like(x)
    for n in range(5):
        taylor_sum += x**n / np.math.factorial(n)
    return taylor_sum

x = np.linspace(-2, 2, 100)
y_exp = exp_function(x)
y_taylor = taylor_series(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y_exp, label=r'$e^x$', color='blue')
plt.plot(x, y_taylor, label=r'Taylor Series', color='red', linestyle='--')

plt.title('Plot of $e^x$ and Taylor Series')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)
plt.show()
```

We see above the function e^x and its close Taylor Series approximation.

# Optimization with Gradient Descent

Gradient descent is an iterative optimization algorithm for finding a local
minimum of a differentiable function.

```python
def gradient_descent(start_x, lr, iters):
    x = [start_x]
    for _ in range(iters):
        grad = der_f(x[-1])
        x.append(x[-1] - lr * grad)
    return np.array(x)

x_vals = gradient_descent(4, 0.1, 20)

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y, label=r'$f(x) = x^2$', color='blue')
plt.scatter(x_vals, f(x_vals), color='red', label='Gradient Descent')

plt.title('Gradient Descent on $f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.show()
```

We see above that gradient descent approaches the minimum.
